Task 2:Preprocessing Function

```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    text = re.sub(r"\d+", "", text)

    text = text.lower()

    text = re.sub(r"(.)\1{2,}", r"\1", text)

    text = re.sub(r"[^\w\s]", "", text)

    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]

    return tokens
```

Task 3:Stress Testing

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product ",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens_list = []

for text in test_sentences:
    tokens = preprocess_text(text)
    cleaned_sentence = " ".join(tokens)

    print("\nOriginal Text:", text)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", cleaned_sentence)

    all_tokens_list.append(tokens)
```

Task 4:Token Analytics

```python
for i, tokens in enumerate(all_tokens_list):
    if len(tokens) == 0:
        total = unique = avg_len = 0
    else:
        total = len(tokens)
        unique = len(set(tokens))
        avg_len = sum(len(t) for t in tokens) / total

    print(f"\nSentence {i+1}")
    print("Total Tokens:", total)
    print("Unique Tokens:", unique)
    print("Average Token Length:", round(avg_len, 2))
```

Task 5:Frequency Analysis

```python
from collections import Counter

all_tokens = [token for sublist in all_tokens_list for token in sublist]

counter = Counter(all_tokens)

print("Top 10 frequent words:")
print(counter.most_common(10))

print("\nTop 5 least frequent words:")
print(counter.most_common()[-5:])
```

Task 6:Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        sentence = " ".join(tokens)

        all_tokens.extend(tokens)
        clean_sentences.append(sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

output = full_pipeline(test_sentences)
print(output)
```

Task 7:Error Handling

```python
import re

def validate_text(text):
    """
    Returns True if text is valid, False otherwise.
    Invalid if:
      - Empty string
      - Only numbers
      - Only emojis / no letters or numbers
      - Not a string
    """
    if not text or not isinstance(text, str):
        return False

    text = text.strip()
    if not text:
        return False

    if text.isdigit():
        return False

    if not re.search(r'[a-zA-Z0-9]', text):
        return False

    return True
```
