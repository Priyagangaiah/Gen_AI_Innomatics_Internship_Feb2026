```python
#Smart Light Controller
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(self.light_name + " is " + self.status)


light = SmartLight("Bedroom Light")
light.turn_on()
light.show_status()
```

```python
# Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def show_id_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)

employee = Employee("Rahul", "EMP102", "IT")
employee.show_id_card()
```

```python
# Mobile Contact Record

class Contact:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone_number)

contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
# Product Price Tag Generator

class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def show_price_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))


# Using the class
item = Product("Headphones", 2499)
item.show_price_tag()
```

```python
# Movie Rating Display System

class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def show_movie_details(self):
        print("Movie:", self.movie_name)
        print("Rating:", self.rating, "/ 5")

movie = Movie("Inception", 4.8)
movie.show_movie_details()
```

```python
# Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def show_delivery_details(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Using the class
order = Delivery("Suman", "Hyderabad")
order.show_delivery_details()
```
