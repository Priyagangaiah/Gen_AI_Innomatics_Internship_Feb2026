```python
# Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    total_likes = 0
    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_post_engagement([200, 300, 250, 300])
```

```python
# Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock_quantity):
    if stock_quantity < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock_quantity)
    print("Status:", status)

check_medicine_stock(6)
```

```python
#Agriculture – Rainfall Adequacy Checker
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0

    for rain in rainfall_data:
        total_rainfall += rain

    average_rainfall = total_rainfall / len(rainfall_data)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average_rainfall)
    print("Rainfall Status:", status)

check_rainfall([70, 75, 80, 63], 70)
```

```python
#Social Media – Duplicate Account Detection
def detect_duplicate_usernames(usernames):
    unique_usernames = set(usernames)

    if len(unique_usernames) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)
detect_duplicate_usernames(["john", "alice", "john"])
```

```python
# Agriculture – Premium Crop Price Filter
def filter_premium_crops(price_list):
    premium_crops = []

    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)

filter_premium_crops([1500, 2500, 1800, 3200])
```

```python
# System Monitoring – Application Health Checker
def check_application_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)

check_application_health(7)
```

```python
# Banking – Daily Transaction Limit Checker
def check_transaction_limit(transaction_amount):
    daily_limit = 50000

    if transaction_amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", transaction_amount)
    print("Transaction Status:", status)

check_transaction_limit(60000)
```

```python
# E-Learning – Student Attendance Eligibility System
def check_exam_eligibility(attendance_record):
    total_days = len(attendance_record)
    days_present = 0

    for day in attendance_record:
        if day == "Present":
            days_present += 1

    attendance_percentage = (days_present / total_days) * 100

    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)

check_exam_eligibility(["Present", "Absent", "Present", "Present", "Present"])
```

```python
# Smart Electricity Bill Analyzer
def calculate_electricity_bill(units_consumed):
    bill = 0

    if units_consumed <= 100:
        bill = units_consumed * 3

    elif units_consumed <= 200:
        bill = (100 * 3) + ((units_consumed - 100) * 5)

    else:
        bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    print("Total Bill:", bill)
    print("Usage Classification:", usage_status)

    return bill, usage_status

calculate_electricity_bill(250)
```
