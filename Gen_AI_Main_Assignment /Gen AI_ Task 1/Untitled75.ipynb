```python
from langchain.prompts import PromptTemplate

# Reusable template
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

def explain_topic(topic):
    return explain_template.format(topic=topic)

# Test
print(explain_topic("Machine Learning"))
```

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def generate_multi_prompt(topic, audience, tone):
    return multi_input_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Test Cases
print(generate_multi_prompt("ML", "beginners", "friendly"))
print(generate_multi_prompt("Python", "kids", "fun"))
print(generate_multi_prompt("Deep Learning", "engineers", "technical"))
```

```python
# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

```python
from langchain.prompts import ChatPromptTemplate

def chat_prompt(role, topic):

    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are an interviewer asking insightful questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Explain {topic}")
    ])

    return chat_template.format(topic=topic)

# Example
print(chat_prompt("teacher", "Neural Networks"))
```

```python
def validate_inputs(audience, tone):

    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("kids", "fun"))  # invalid audience → default applied
```

```python
def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} teaching style"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Ask interview questions about {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} storytelling style"
        )

    else:
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} for {audience} in a {tone} tone"
        )

    return template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

# Example
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
# ONE reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 Different Inputs
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "casual"),
    ("Deep Learning", "expert", "formal"),
    ("Data Science", "beginner", "casual"),
    ("NLP", "expert", "formal")
]

for topic, audience, tone in inputs:
    print(reusable_template.format(
        topic=topic,
        audience=audience,
        tone=tone
    ))
```
